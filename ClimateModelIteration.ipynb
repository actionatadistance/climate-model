# You Can Prove Global Warming
Worldwide average temperatures have been increasing since industrialization.  The scientific consensus is that the significant majority of this temperature shift is from human production of greenhouse gas - predominantly CO2.  The models that imply this effect have been understood by physical chemists since we came to understand the nature of light's interaction with matter.

The conceptual model of climate is quite simple.  Sunlight strikes earth warming it.  The CO2 acts in place of the glass of a greenhouse helping that warmth persist.  Our activities in extracting carbon based fuels and burning them for energy leads to increased atmospheric CO2.  Adding more CO2 leads to increased greenhouse effect and so higher average temperatures.

Despite the simplicity of the model there is widespread skepticism about it.  I think a large portion of this skepticism is because from the 'climate change exists' side the argument is largely one from authority.  'The scientists say climate change exists so it must be true!' There are many reasonable rebuttals to this, and merely parroting the scientists and pointing at temperature measurements for the last couple of decades is insufficient.

The points of contention in climate modelling are:

- How much (if any) does atmospheric CO2 induce warming?
- How much has the global average temperature changed since we have produced this CO2?

# How Much Of The Atmospheric CO2 Is From Human Activity

This is an interesting question. At first it can seem relatively easy - just count up all of the CO2 we produce and compare it to the CO2 in the atmosphere.  Then we need to take into account the natural processes that distribute CO2: absorption into the ocean surface and associated processes bringing the CO2 into the deep ocean, photosynthetic respiration, volcanic emissions, stone weathering.  We will go over the numbers involved.

## Total Atmospheric CO2

Atmospheric CO2 content can be measured at home - at least at ground level.  Calculating total atmospheric CO2 is fairly simple as atmospheric CO2 concentration by altitude is publicly available data, so computing the total mass of atmospheric CO2 from this is fairly trivial.

## Human CO2 Production

This is quite straightforward to compute.  We have fairly well sourced publications reporting energy usage per each fossil fuel by year back to 1800.  I will be using the [Our World In Data](https://ourworldindata.org/fossil-fuels) dataset for consumption per fossil fuel.  After that we just compare energy usage to per-unit CO2 emissions which we can calculate ourself from the relevant combustion equations, as well as use publications reporting the relative emissions.

Concrete production is also a source of CO2 production and is a short-run CO2 producer, aside from the fossil fuels used in the process and accounted for above concrete is net neutral.  As concrete weathers it re-absorbs the CO2 released in its production.  This does mean that we should be properly disposing of waste concrete rather than burying it in a landfill where this weathering occurs slowly but concrete is a small minority CO2 producer.

## Oceanic CO2 Absorption

Since only atmospheric CO2 is relevant to global warming and since CO2 is absorbed into the ocean if the ocean was able to just absorb it all that would be great.  There are two processes by which CO2 is distributed within the ocean, diffusion and ocean turnover.

Diffusion is the process when you put a bit of food dye in a cup and the dye spreads throughout the cup. Diffusion is effective at distributing CO2 close to the ocean surface.  The limiting factor that prevents diffusion from being effective in distributing to the deep ocean is that the CO2 concentration in water is limited via chemical means and the two-way nature of diffusion means that the process ends up at a near equilibrium leading to a slow distributive process - but 'near equilibrium' and 'slow distributive' are just descriptors - we will compute approximate numbers for annual sequestration from these processes.

Ocean turnover is the process by which surface waters of the ocean get pushed down into the deep sea by various ocean currents. On a per-liter involved basis this is the most efficient way to distribute water in the ocean as a liter turned over is essentially immediately acting out the entire process of diffusion.  The problem is that only a small amount of the oceans water turns over, as we can see from the existing haloclines (differences in salinity by depth).

## Photosynthetic Respiration

Photosynthetic respiration involves plants and algae taking CO2 from the air and combining it with water and turning it into sugars used to build up their biomass.  With this we can see plants/algae are a possible place to put this excess CO2.  We can actually see that photosynthesis out-consumes our CO2 production - the problem is that when the organism dies it rots and the CO2 is released.  We can see high-definition atmospheric CO2 graphs that there's an annual peak of atmospheric CO2 in the Northern Hemisphere winter and an annual trough in its summer.  The net effect is largely in balance - but we can confirm that.

## Volcanic Activity

Volcanoes can, depending on the magma content, emit CO2.  The amount is detectable and monitored.

## Stone Weathering

Olivine stone chemically reacts with CO2 during its weathering (chemical decomposition) processes.  Since solids (rock) don't readily allow CO2 to penetrate this ends up suffering from the same problems as oceanic diffusion - but it is a relevant CO2 sink.


# How Much (If Any) Does Atmospheric CO2 Induce Warming?

This is the most technically challenging portion of the process.  It only involves two processes - light absorption and light emission - but each of these processes is complex.

When light radiation approaches earth it first passes through the atmosphere then hits the earth's surface (or a cloud) at which point it is either reflected or absorbed.  All of the energy from absorbed light turns to heat.  This heat is then disipated by the earth emitting radiation (predominantly infra-red radiation).  Some of this radiation is then re-absorbed greenhouse gasses.  These two processes in a static atmosphere will come to an equilibrium temperature.  The way to find the equilibrium temperature is fairly straightforward.

## Earth's Energy Balance

To find the equilibrium temperature in a static atmosphere, we define the following variables:

- Let \( E_s \) represent the light energy absorbed from the sun,
- \( E_e \) represent the light energy emitted from the earth, and
- \( E_a \) represent the light energy absorbed by the atmosphere.

Thus, we have the equation:

$$
E_s = E_e - E_a
$$

Awkwardly none of these terms are actually temperature - so solving for temperature is frustrated.  However - E_e and E_a are functionally dependent upon temperature, so we need to go technically deeper into these terms.

## Black Body Radiation

If you take any piece of matter - some rock, the Sun, the Earth, yourself, etc - that has a temperature above absolute zero that piece of matter is continuously emitting radiation according to Planck's Law.  Planck's law identifies how bright an object is at a given frequency given that object's temperature.  For example - 

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define physical constants
h = 6.626e-34  # Planck's constant in Js
c = 3.00e8     # Speed of light in m/s
k_B = 1.381e-23  # Boltzmann constant in J/K

# Define Planck's Law function for spectral radiance by frequency
def planck_law_frequency(nu, T):
    """Calculate the black body spectral radiance at a given frequency and temperature."""
    return (2 * h * nu**3 / c**2) / (np.exp(h * nu / (k_B * T)) - 1)

# Set temperature in Kelvin
T = 3000  # Temperature of the black body, you can change this value

# Define frequency range (in Hz)
frequencies = np.linspace(1e12, 5e14, 1000)  # Range from 1 THz to 300 THz

# Calculate spectral radiance for each frequency
intensities = planck_law_frequency(frequencies, T)

# Plotting the result
plt.figure(figsize=(10, 6))
plt.plot(frequencies, intensities, color="blue")
plt.title(f"Black Body Spectral Radiance at {T} K")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Spectral Radiance $B(\\nu, T)$ (Wm$^{-2}$Hz$^{-1}$sr$^{-1}$)")
plt.grid(True)
plt.show()
```
